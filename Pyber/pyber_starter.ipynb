```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# File to Load (Remember to change these)
city_data_to_load = os.path.join( "data/city_data.csv")
ride_data_to_load = os.path.join("data/ride_data.csv")

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merge_df = pd.merge(ride_data, city_data, on="city", how="left")
# Display the data table for preview
merge_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_city_type = merge_df[merge_df["type"] == "Urban"].groupby([merge_df["city"]])
rural_city_type = merge_df[merge_df["type"] == "Rural"].groupby([merge_df["city"]])
suburban_city_type = merge_df[merge_df["type"] == "Suburban"].groupby([merge_df["city"]])
# calculate urban ride ID(x), fare average(y), and driver count avg(z)
x_urban = urban_city_type["ride_id"].count()
y_urban = urban_city_type["fare"].mean()
z_urban = urban_city_type["driver_count"].mean()

# calculate rural ride ID(x), fare average(y), and driver count avg(z)
x_rural = rural_city_type["ride_id"].count()
y_rural = rural_city_type["fare"].mean()
z_rural = rural_city_type["driver_count"].mean()

# calculate suburban ride ID(x), fare average(y), and driver count avg(z)
x_suburban = suburban_city_type["ride_id"].count()
y_suburban = suburban_city_type["fare"].mean()
z_suburban = suburban_city_type["driver_count"].mean()

# Build the scatter plots for each city types
plt.scatter(x_urban, y_urban, label = "Urban", s=z_urban * 10, color=["coral"], edgecolor="black", alpha = 0.75, marker="o")

plt.scatter(x_rural, y_rural, label = "Rural", s=z_rural * 10, color=["gold"], edgecolor="black", alpha = 0.75, marker="o")

plt.scatter(x_suburban, y_suburban, label = "Suburban", s=z_suburban  * 10, color=["lightskyblue"], edgecolor="black", alpha = 0.75, marker="o")

plt.grid()

# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2018)")

# Create a legend
legend = plt.legend(fontsize = 8, title= "City Types", loc="best")
# make legend handles
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(42,35,"Note: \nCircle size correlates with driver count per city.", fontsize = 10)

# Save Figure
plt.savefig("Images/RideSharingData.png", bbox_inches="tight")
```

```python
# Show plot
plt.show()
```
